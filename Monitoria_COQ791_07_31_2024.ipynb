```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def Conc(y, t, D, dx, cs, nodes):
    dcdt = [0] * nodes
    dcdt[0] = (D / (dx ** 2)) * ((2 / 3) * y[1] - (2 / 3) * y[0] - (2 / 3) * cs * dx)
    for i in range(1, nodes - 1):
        dcdt[i] = (D / (dx ** 2)) * (y[i + 1] - 2 * y[i] + y[i - 1])
    dcdt[nodes - 1] = (D / (dx ** 2)) * (cs + y[nodes - 2] - 2 * y[nodes - 1])
    return dcdt

D = 10
L = 15
c0 = 12
cs = 5
nodes = 200
dx = L / nodes

t = np.linspace(0, 1, 200)
y0 = [c0] * nodes

params = (D, dx, cs, nodes)
sol = odeint(Conc, y0, t, args=params)
```

```python
print(sol[:,-1])
```

```python
plt.plot(t,sol[:,-1],'r')
plt.title('Concentração pelo Tempo')
plt.xlabel('t')
plt.ylabel('Concentração')
plt.show()
```

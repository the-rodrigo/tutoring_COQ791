```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def Duffing(S, t):
    x1, x2 = S
    dx1dt = x2
    dx2dt = x1 - x2 - x1**3
    return [dx1dt, dx2dt]
```

```python
t = np.linspace(0, 20, 100)
x10 = np.linspace(-3,3,100)
x20 = np.linspace(-3,3,100)

for i in range(len(x10)):
    for j in range(len(x20)):

        S0 = [x10[i], x20[j]]
        sol = odeint(Duffing, S0, t)

        x1_plot = sol[:, 0]
        x2_plot = sol[:, 1]

        if round(x1_plot[-1]) > 0:
            plt.scatter(x10[i], x20[j], color='r')
        else:
            plt.scatter(x10[i], x20[j], color='b')

plt.scatter(0, 0, color='w')
plt.scatter(-1, 0, color='w')
plt.scatter(1, 0, color='w')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.title('Plano de Fases de x_1 vs x_2')
plt.show()
```
